```python
import pandas as pd
```

```python
df_pagamento = pd.read_csv(r"C:\Users\nada_\OneDrive\Desktop\REPROGRAMA\on33-python-S14-analise-de-dados\exercicios\para-casa\Francitelma\olist_order_payments_dataset.csv")
df_pagamento

```

```python
info_df = df_pagamento.info()
print(info_df)
```

```python
def visualizar_as_duplicadas(df_pagamento):  #sem duplicadas
    duplicados = df_pagamento[df_pagamento.duplicated(keep=False)]

    return duplicados

linhas_duplicadas = visualizar_as_duplicadas(df_pagamento)
print (linhas_duplicadas)
```

```python
df_pagamento.describe(include='all')
```

```python

# Contar o número de pagamentos por order_id
pagamentos_por_pedido = df_pagamento.groupby('order_id').size().reset_index(name='num_pagamentos')

# Filtrar pedidos com múltiplos pagamentos
pedidos_multiplos_pagamentos = pagamentos_por_pedido[pagamentos_por_pedido['num_pagamentos'] > 1]

print(pedidos_multiplos_pagamentos.head(200000))
```

```python

# Identificar order_ids com múltiplos registros
contagem_order_id = df_pagamento['order_id'].value_counts()

# Filtrar apenas os pedidos com mais de um registro
multiplos_registros = contagem_order_id[contagem_order_id > 1]

print(f"Número de pedidos com múltiplos registros: {len(multiplos_registros)}")

# Exibir alguns exemplos de pedidos com múltiplos registros
print(multiplos_registros.head())

# Consolidação dos pagamentos por pedido
# Agrupar por order_id e somar os valores de pagamento
consolidado_pagamento = df_pagamento.groupby('order_id').agg({
    'payment_value': 'sum',
    'payment_installments': 'sum',  # Ou usar 'max', dependendo do contexto
    'payment_type': lambda x: ', '.join(x.unique())  # Lista os tipos de pagamento usados
}).reset_index()

print(consolidado_pagamento.head())
```

```python
# Contar o número de pagamentos por pedido
pagamentos_por_pedido = df_pagamento.groupby('order_id').size().reset_index(name='num_pagamentos')

# Contar a frequência de cada número de pagamentos
pagamentos_frequencia = pagamentos_por_pedido['num_pagamentos'].value_counts().sort_index()

# Visualizar pagamentos_frequencia
print(pagamentos_frequencia)

                                        #Número de pedidos com múltiplos registros: 2961
```
